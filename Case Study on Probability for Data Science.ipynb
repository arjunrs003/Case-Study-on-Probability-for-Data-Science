We can see that gill_attachment and veil_color are highly correlated to each other. We can remove veil_color from our dataset. 

```python
df.drop('veil_color',inplace=True,axis=1)
```

```python
#Let us split our dataset into x and y
x=df.drop('class',axis=1)
y=df['class']
```

```python
#Let us split train and test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

# Modelling

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
Gaussian_nb=GaussianNB()
Gaussian_nb.fit(x_train,y_train)
```

```python
#let us predict on our test data
y_nb_pred=Gaussian_nb.predict(x_test)
```

```python
#let us check our accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

print("Accuracy is: ", accuracy_score(y_test,y_nb_pred))
print("Precision is: ", precision_score(y_test,y_nb_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_nb_pred, average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_nb_pred, average = 'micro'))
```

```python
from sklearn.naive_bayes import BernoulliNB
bernoulli_nb=BernoulliNB()
bernoulli_nb.fit(x_train,y_train)
```

```python
y_bernoulli_nb_pred=bernoulli_nb.predict(x_test)
```

```python
#let us check our accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

print("Accuracy is: ", accuracy_score(y_test,y_bernoulli_nb_pred))
print("Precision is: ", precision_score(y_test,y_bernoulli_nb_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_bernoulli_nb_pred, average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_bernoulli_nb_pred, average = 'micro'))
```

# Other models

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
classifiers=[('LogisticRegression',lr),('KNeighbors',knn),('DecisionTree',dt)]
for clf_name,clf in classifiers:
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    print("Accuracy for: ",clf_name, accuracy_score(y_test,y_pred))
    print("Precision is: ",clf_name, precision_score(y_test,y_pred))
    print("Recall is: ",clf_name, recall_score(y_test,y_pred))
    print("F1 score is: ",clf_name, f1_score(y_test,y_pred))
```
